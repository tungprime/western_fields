Here are some functions that I wrote to produce the graphs that I introduced in the lecture.

```python
def complete_graph(n):
    g=nx.empty_graph()
    for i in range(0,n):
        for j in range(0, n):
            if i !=j: 
                g.add_edge(i,j)
    return g      

def cycle_graph(n):
    g=nx.Graph()
    for i in range(0,n-1):
        g.add_edge(i, i+1)
    g.add_edge(n-1, 0)    
    return g    

def paley(q):
    squares = set([(a**2)%q for a in range(1,q)])
    g=nx.empty_graph(q)
    for i in range(0,q):
        for j in range(0,q):
            if (i-j)%q in squares:
                g.add_edge(i,j)
    return g  

def ring_graph(N,k):
    g = nx.empty_graph(N)
    for i in range(0,N):
        for j in range(0,N):
            if (i-j)% N !=0 and (i-j)%N <=k:
                g.add_edge(i,j)
    return g            
                
```

```python
#the complete graph with N=6 nodes
fig = plt.gcf()
fig.set_size_inches(8,6)
g = complete_graph(6)
nx.draw_circular(g, with_labels = True,
                node_size = 600,
                node_color = '#6495ED')
plt.savefig("complete_graph.png", dpi =300)
```

```python
# the cycle graph with N=6 nodes.
fig = plt.gcf()
fig.set_size_inches(8,6)
g = cycle_graph(6)
nx.draw_circular(g, with_labels = True,
                node_size = 600,
                 scale = 2,
                node_color = '#6495ED')
plt.savefig("cycle_graph.png", dpi =300)
```

```python
# Paley graph for q=13 
fig = plt.gcf()
fig.set_size_inches(8,6)
g = paley(13)
pos = nx.circular_layout(g)
nx.draw(g, pos=pos,
       with_labels = True,
                node_size = 600,
                 scale = 2,
                node_color = '#6495ED')
plt.savefig("paley_graph_13.png", dpi =300)
```

```python
#Paley graph for q=5
fig = plt.gcf()
fig.set_size_inches(8,6)
g = paley(5)
pos = nx.circular_layout(g)
nx.draw(g, pos=pos,
       with_labels = True,
                node_size = 600,
                 scale = 2,
                node_color = '#6495ED')
plt.savefig("paley_graph_5.png", dpi =300)
```

```python
# Ring graph for N=7, k =2
fig = plt.gcf()
fig.set_size_inches(8,6)
N=7
k=2
g = ring_graph(N,k)
pos = nx.circular_layout(g)
nx.draw(g, pos=pos,
       with_labels = True,
                node_size = 600,
                 scale = 2,
                node_color = '#6495ED')
plt.savefig("ring_graph.png", dpi =300)
```
