Here are some functions that I wrote to produce the graphs that I introduced in the lecture.

```python
def complete_graph(n):
    g=nx.Graph()
    for i in range(1,n+1):
        for j in range(1, n+1):
            if i !=j: 
                g.add_edge(i,j)
    return g      

def cycle_graph(n):
    g=nx.Graph()
    for i in range(1,n):
        g.add_edge(i, i+1)
    g.add_edge(n, 1)    
    return g    

def paley(q):
    squares = set([(a**2)%q for a in range(1,q)])
    g=nx.empty_graph(q)
    for i in range(0,q):
        for j in range(0,q):
            if (i-j)%q in squares:
                g.add_edge(i,j)
    return g            
                
```

```python
g = complete_graph(6)
nx.draw_circular(g, with_labels = True,
                node_size = 600,
                node_color = '#6495ED')
plt.savefig("complete_graph.png", dpi =300)
```

```python
g = cycle_graph(6)
nx.draw_circular(g, with_labels = True,
                node_size = 600,
                 scale = 2,
                node_color = '#6495ED')
plt.savefig("cycle_graph.png", dpi =300)
```

```python
g = paley(13)
pos = nx.circular_layout(g)
nx.draw(g, pos=pos,
       with_labels = True,
                node_size = 600,
                 scale = 2,
                node_color = '#6495ED')
plt.savefig("paley_graph.png", dpi =300)
```
